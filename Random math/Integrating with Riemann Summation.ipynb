```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

e = np.exp(1)

def f(x):
    return (-1)*(x**2) - 2 # (1/2)*(x**3) - x + 2

a = -2 # float(input('You are integrating the fuction (x^3)/2 - x + 2. Where do you want to start the integration? '))
b = 2 # float(input('You are integrating the fuction (x^3)/2 - x + 2. Where do you want to finish the integration? '))
h = 0.001 # float(input('You are integrating the fuction (x^3)/2 - x + 2. What do you want to be your step size? '))

xvalues = []
yvalues = []

def integrate(f, a, b):

    n = (b-a)/h
    integral = 0
    for i in range(int(n)):
        y = f(a + (i*h))
        yvalues.append(y)
        xvalues.append(a + (i*h))
        dArea = ((xvalues[i] - (xvalues[i]-h)) * yvalues[i])
        integral = integral + dArea
    return yvalues, xvalues, integral 

values = integrate(f, a, b)

print ("Integral é ", values[2])

plt.plot(xvalues, yvalues)
plt.grid()
#plt.show
        
    
    
```
