```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (4,4)


n = 8 #Power which you want to elevate x and y to to get the super egg. X^n + Y^n = 1
n1 = n
n2 = n

x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**n1 + Y**n2
plt.contour(X,Y,F,[1])
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.show()
```
