```python
#Importing cell

import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
#Input cell

x = np.linspace(0, 35, 1000) #Number of occurrences of some event. X axis
lambdas = [1, 5, 8, 12, 21] #list of values of lambda (could be only one)
```

```python
#Functions cell.

'''Poisson distribution: P(x=k) = (e^(-\lambda)*\lambda**k)/k!. Here, k is the values in the x-axis,
and lambda is some lambda in the list on the cell above'''
def poisson(l,k): #Defining Poisson distribution
    y = (l**k)/(np.exp(l)*scipy.special.factorial(k))
    return y
```

```python
#Plotting cell

'''Here i plot the graphs of the poisson distribution for different values os \lambda'''
for l in lambdas:
    plt.plot(x, poisson(l, x), label = '$\lambda$ = ' + str(l))

plt.axhline(linewidth=1, color='k') #Highlighting x-axis
plt.axvline(linewidth=1, color='k') #Highlighting y-axis

plt.xlabel("Number of occurrences")
plt.ylabel('Probability of k occurrences for a given $\lambda$')

plt.title('Poisson distribution for different values of $\lambda$')
plt.grid()
plt.legend()
plt.show()
```
