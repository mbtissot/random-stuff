```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3
b = 2
mult = a/b


A = 1
B = 1
omegax = 1
omegay = mult*omegax
alpha=0
beta=0
t = np.arange(0, 50, 0.1)

xvals = []
yvals = []

frac = f'\\frac{{{a}}}{{{b}}}'
#\print(frac)
#'''#dê um print(frac) p conferir oq isso faz. mas só cria uma string com \frac{valorDe'a'}{valorDe'b'}'''
```

```python
def x(t):
    return A*np.sin(omegax*t - alpha)

def y(t):
    return B*np.cos(omegay*t - beta)
```

```python
for i in range(len(t)):
    xvals.append(x(t[i]))
    yvals.append(y(t[i]))
```

```python
plt.plot(xvals, yvals, 'r', label=r'$\omega_y = '+ frac + '\omega_x$')
plt.plot([-1.1, 1.1], [0, 0], 'k') #'eixo' x
plt.plot([0, 0], [-1.1, 1.1], 'k') #'eixo' y
plt.xlim(-1.05, 1.05)
plt.ylim(-1.05, 1.05)
plt.xlabel(r'X')
plt.ylabel(r'Y')
plt.legend(prop={'size': 12})
plt.grid()
plt.show()
```
